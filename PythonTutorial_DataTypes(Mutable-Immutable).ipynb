# **Data types in Python:**

Different data types in Python, including integers, floating-point numbers, strings, booleans, lists, tuples, and dictionaries, etc.

## 1. **Integers:**

```python
1. Integers:
Integers represent whole numbers, both positive and negative, without any decimal points. For example:
"""

x = 5
y = -10
"""

print("x: ", type(x))
print("y: ", type(y))
```

## 2. **Floating-Point Numbers:**

```python
"""
2. Floating-Point Numbers:
Floating-point numbers, also known as floats, represent numbers with decimal points. For example:

"""

a = 3.14
b = -2.5

print("a:", type(a))
print("b:", type(b))

print(id(a))
print(id(b))
```

```python
a = 5
b = 5.6
sum = a + b

print(type(sum))
```

```python
s1 = "35"
s2 = "35rtr"

print(int(s1))

int(s2)
```

## 3. **Strings:**

```python
"""
3. Strings:
Strings are sequences of characters enclosed in single or double quotation marks. They are used to represent text. For example:
"""

name = "John"
message = 'Hello, world!'

print(type(name))
print(type(message))
```

```python
# string examples:
s1 = "Python"
s2 = " is a Programming language"

out = s1 + s2
print(out)

print(id(s1))
print(id(s2))
print(id(out))
```

## 4. **Booleans:**

```python
"""
4. Booleans:
Booleans are a data type that represents either true or false. They are often used in conditional statements and logical operations. For example:
"""

is_true = True
is_false = False

print(type(is_true))
print(type(is_false))
```

# **Mutable Data Types:**
* we can modify the data after defining it.
* Mutable Data Types: List, Dict, Set

# **Immutable Data Types:**
* We can not do any modification after defining it.
* Except these 3 (list, dict, set), everything is immutable data types in python.

## 5. **Lists:**

```python
"""
5. Lists:
Lists are ordered collections of items, enclosed in square brackets ([]). They can store elements of different data types and are mutable, which means you can modify them.
Lists can contains haterogenous data.

For example:
"""

numbers = [1, 2, 3, 4, 5, 5.6, "name", [5]]

names = ["Alice", "Bob", "Charlie"]
mixed_list = [10, "hello", True]

print(type(numbers))

```

## 6. **Tuples:**

```python
"""
6. Tuples:
Tuples are similar to lists (ordered), but they are enclosed in parentheses (()). Unlike lists, tuples are immutable, meaning their elements cannot be changed after creation. For example:
"""

point = (3, 4)
colors = ("red", "green", "blue")

print(type(point))
```

```python
# Tuple behaviour with single item
a = ()
b = (5.678)
c = (6,)

print(type(a))
print(type(b))
print(type(c))
```

## 7. **Dictionaries:**

```python
"""
7. Dictionaries:
Dictionaries are unordered collections of key-value pairs, enclosed in curly braces ({}). Each element in a dictionary is accessed by its key rather than by an index. Keys must be unique, and they are used to retrieve their corresponding values. For example:
"""

person = {"name": "Alice", "age": 25, "city": "New York"}

print(type(person))

person['name']
```

```python
dict_2 = {'id': 5, "roll_no": 5}

dict_2['id']
dict_2['roll_no']
```

```python
dict_3 = {"name": 'ekta', 'name': 'anjana', 'name': 'jyotsna'}

dict_3['name']
```
