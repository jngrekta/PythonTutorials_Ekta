# **String Data Type:**

## 1. **String Concatenation:**
String concatenation is the process of combining two or more strings together.

```python
string1 = "Hello"
string2 = "World"
concatenated_string = string1 + " " + string2
print(concatenated_string)  # Output: Hello World
```

## 2. **String Length:**
The len() function is used to determine the length (number of characters) of a string.

```python
string = "Hello World  "
length = len(string)
print(length)  # Output: 11
```

## 3. **String Indexing and Slicing:**
String indexing is used to access individual characters in a string, while string slicing is used to extract a portion of a string.

```python
string = "Hello World"
first_character = string[4]
print(first_character)  # Output: H
```

```python
s = "Python" # 0, 1, 2, 3, 4, 5

print(s[5])

print(s[6])
```

```python
# indexing:
# +ve indexing: left to right (0, 1, 2,...)
# -ve indexing: right to left (-1, -2, -3,......)

s = "Hello World"

# 0, 1, ,2 ,3, ....

print(s[0:len(s):1])

# negative indexing:

print(s[-3])

print("*************")
print(s[4::1])
print(s[4::-1])

```

```python
# negative indexing in slicing

name = "Ekta Anjana Jyotsna"

print(name[10:4:-1])

print(name[18:11:-1])

print(name[3:0:-1])

print(name[3::-1])

print(name[::-1])
```

```python
# string slicing

string = "Python is a programming language."



print(len(string))

print(string[10])

print(string[:11])

print(string[:])

print(string[12:23])

print(string[0:6:2]) # P, t, o

print(string[12:23:3]) # p, g, m, n




```

## 4. **String Formatting:**
String formatting allows you to insert dynamic values into a string using placeholders or format specifiers.

```python
name = input("Enter your name: ")
age = input("Enter your age: ")
city = input("Enter your city name: ")

string = "My name is {} and I am {} years old. I am living in {}"
formatted_string = string.format(name, age, city)

print(formatted_string)
```

```python
height = "5'1"
weight = "60"

string = "My height is {} and my weight is {}.".format(height, weight)

print(string)
```

```python
# formatting with f-strings
name = "Ram"
language = "English"

string = f"My name is {name}, I can speak {language} langauge."
print(string)
```

```python
maths = 60
science = 70

string2 = f"maths marks {maths}, science marks {science}."
print(string2)
```

# **String Methods:**
Python provides various built-in string methods to perform operations on strings, such as converting case, finding substrings, splitting, and joining strings, etc.

### 1. **len()** - Returns the length of a string.

```python
string = "Hello, world!"
length = len(string)
print(length)  # Output: 13
```

### 2. **lower()** - Converts a string to lowercase.

```python
string = "HellO, woRld!"
lowercase_string = string.lower()
print(lowercase_string)  # Output: hello, world!
```

### 3. **upper()** - Converts a string to uppercase.

```python
string = "Hello, world!"
uppercase_string = string.upper()
print(uppercase_string)  # Output: HELLO, WORLD!
```

### 4. **capitalize()** - Converts the first character of a string to uppercase and the rest to lowercase.

```python
string = "hello, World!"
capitalized_string = string.capitalize()
print(capitalized_string)  # Output: Hello, world!
```

### 5. **title()** - Converts the first character of each word in a string to uppercase and the rest to lowercase.

```python
string = "python prograMming languaGE"
title_string = string.title()
print(title_string)  # Output: Hello, World!
```

### 6. **strip()** - Removes leading and trailing whitespace from a string.

```python
string = "        Hello, world!   "
stripped_string = string.strip()
print(stripped_string)  # Output: Hello, world!
```

```python
string = "    Ram is eating food. Hello,     world!         gusdgshdjghdjzbghjzhjz          "
stripped_string = string.strip()
print(stripped_string)  # Output: Hello, world!
```

### 7. **split()** - Splits a string into a list of substrings based on a specified separator.

```python
string = "Hello world python programmig"
split_string = string.split(" ") # default separtor is space " "
print(split_string)

type(split_string)
```

```python
string = "Ekta Jyotsna.  Anjana "
split_string = string.strip().split(" ")
print(split_string)

type(split_string)
```

### 8. **join()** - Concatenates elements of an iterable (e.g., a list) into a single string using a specified separator.

```python
list_of_words = ['Ekta', 'Jyotsna.', '', 'Anjana']
joined_string = "-".join(list_of_words)
print(joined_string)  # Output: Hello, world!

type(joined_string)
```

```python
subjects = ['maths', 'science', 'history']

joined_string = " ".join(subjects)
print(joined_string)
```

### 9. **find()** - Returns the index of the first occurrence of a substring in a string.

```python
string = "Hello World python World"
substring_index = string.find("World")
print(substring_index)  # Output: 6
```

```python
string = "Hello world python World"
substring_index = string.find("World")
print(substring_index)
```

### 10. **count()** - Returns the number of occurrences of a substring in a string.

## **'in' and 'not in' operators with strings:**

### 1. **Using 'in' operator:**

```python
string = "Hello, World!"

# Check if a substring is present in the string
print("Hello" in string)  # Output: True

# Check if a character is present in the string
print("o" in string)  # Output: True

# Check if a substring is not present in the string
print("Python" in string)  # Output: False
```

```python
string_2 = "Ram Sita Mohan"

word = 'Ram'
if 'Rama' in string_2:
  print(f"{word} is there in the string.")
else:
  print(f"{word} is not present in string")
```

### 2. **Using not in operator:**

```python
string = "Hello, World!"

# Check if a substring is not present in the string
print("Python" not in string)  # Output: True

# Check if a character is not present in the string
print("z" not in string)  # Output: True

# Check if a substring is present in the string
print("Hello" not in string)  # Output: False
```

# **String Unicode to Character Conversion and Character to Unicode Conversion in Python:**

Unicode is a universal character encoding standard that assigns a unique number (code point) to every character from all writing systems. In Python, you can convert between Unicode and characters using the **chr()** and **ord()** functions.

## **Unicode to Character Conversion:**
The **chr()** function in Python takes a Unicode code point as input and returns the corresponding character.

```python
# Here's an example:

unicode_code_point = 67
character = chr(unicode_code_point)
print(character)  # Output: 'C'

```

## **Character to Unicode Conversion:**
The **ord()** function in Python takes a character as input and returns its corresponding Unicode code point.

```python
# Here's an example:

character = 'A'
unicode_code_point = ord(character)
print(unicode_code_point)  # Output: 65
```

## **Unicode Limits for Alphabets:**
In Python, the lowercase and uppercase alphabets have specific Unicode ranges.
* The Unicode values for lowercase letters **'a' to 'z'** range from **97 to 122**.
* For uppercase letters **'A' to 'Z'** range from **65 to 90**.



```python
# The below ranges can be useful when working with string manipulations.

print(ord('a'))
print(ord('z'))
```

```python
# example:

lowercase_start = ord('a')
lowercase_end = ord('z')
uppercase_start = ord('A')
uppercase_end = ord('Z')

print(f"Lowercase Unicode range: {lowercase_start}-{lowercase_end}")
print(f"Uppercase Unicode range: {uppercase_start}-{uppercase_end}")
```
